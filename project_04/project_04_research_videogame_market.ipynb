# Изучение закономерностей, определяющих успешность игр

Мы работаем в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation).

**Цель исследования**:

Необходимо выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.
    

**Ход исследования**:

Перед нами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и мы планируете кампанию на 2017-й. Нужно отработать принцип работы с данными. Неважно, прогнозируем ли мы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.

В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

Для подсчета количества и процента пропущенных значений создадим функцию `df_statistic`

```python
def statistic(df):
    df_stat = pd.concat([df.isna().sum(), df.isna().mean().map('{:.2%}'.format)], axis=1)
    df_stat.columns = ['количество пропущенных значений', 'процент пропущенных значений']
    return df_stat
```

Напишем функцию investigate_column для анализа данных в колонке таблицы

```python
def investigate_column(df, column, is_category):
    '''
    аргументы функции:
    df - имя таблицы
    column - имя колонки датафрейма
    is_category - True для категориального признака
    '''
    display(statistic(df).query('index == @column'))
    if not is_category:
        print('')
        print('Метод describe:')
        print(df[column].describe())
    else:
        print('')
        print('Метод value_counts:')
        print(df[column].value_counts())
```

### Рассчет суммарных продаж по всем регионам

Создадим колонку с общемировым значением продаж `total_sales`:

```python
df['total_sales'] = df['na_sales'] + df['eu_sales'] + df['jp_sales'] + df['other_sales']
```

**Продолжим анализ появления пропущенных значений в датасете:**

Напишем функцию `set_game_region`, которая будет возвращать имя региона, в котором доля продаж игры превышает 75% общемировых продаж: если такой регион есть, то назовем его `домашний регион` для конкретной игры. Если такого региона для игры нет, то функция вернет `global`

```python
def set_game_region(row):
    sales = [row.na_sales, row.eu_sales, row.jp_sales, row.other_sales]
    names = ['na', 'eu', 'jp', 'other']
    for region, name in zip(sales, names):
        if region:
            if (region / row.total_sales) >= 0.75:
                return name
    return 'global'
```

С помощью функции `set_game_region` заполним колонку `game_region`:

```python
df['game_region'] = df.apply(set_game_region, axis=1)
```

Посмотрим на распределение игр по домашнему региону `game_region`:

```python
game_region_share = df['game_region'].value_counts().apply(lambda x: x / len(df))
print(game_region_share)
```

```python
rating_na_stat = (
    df
    .groupby('game_region')['rating']
    .agg([lambda x: x.isna().sum()/len(x), lambda x: x.isna().sum()])
    .join(game_region_share)
    .rename({'<lambda_0>': '% пропущенных значений в регионе', 
             '<lambda_1>': 'Количество пропущенных значений в регионе', 
             'game_region': '% от общемирового количества игр'}, axis=1)
    .rename(index={'eu': 'Европа', 'global': 'Весь мир', 'jp': 'Япония',\
                   'na': 'Северная Америка', 'other': 'Другие страны'})
    .style.set_caption('Анализ пропущенных значений в колонке RATING')
    .format({'% пропущенных значений в регионе': '{:.2%}', '% от общемирового количества игр': '{:.2%}'})
    .background_gradient(cmap='rocket_r', axis=0, 
                         subset=['% пропущенных значений в регионе', '% от общемирового количества игр'])
)
rating_na_stat.index.name = 'Домашний регион игры'
display(rating_na_stat)
```

Значит компания ESRB вообще не интересует японцев

### Самые популярные платформы (топ-5)

Напишем функцию `top_statistic`, которая возвращает ТОП5 сгруппированной таблицы  по колонке, которую мы ей передаем:

```python
def top_statistic(df, groupby_column, agg_column):
    """
    df - датафрейм
    groupby_column - колонка, по которой происходит группировка
    agg_column - колонка по которой происходит агрегирование
    """
    df_top = (
        df
        .groupby(groupby_column)
        .agg({agg_column: 'sum'})
        .sort_values(by=agg_column, ascending=False)
        .head()
    )
    return df_top
```

Вызовем функцию `top_statistic`, чтобы получить ТОП5 по продажам в северной америке с группировкой по платформе:

```python
top_platform_na = top_statistic(actual_games, 'platform', 'na_sales')
display(top_platform_na)
```

Вызовем функцию `top_statistic`, чтобы получить ТОП5 по продажам в европе и японии с группировкой по платформе:

```python
top_platform_eu = top_statistic(actual_games, 'platform', 'eu_sales')
top_platform_jp = top_statistic(actual_games, 'platform', 'jp_sales')
```

Нарисуем графики с долями продаж по платформам для каждого региона

Напишем функцию `top_statistic_v2`, которая возвращает ТОП5 сгруппированной таблицы  по колонке, которую мы ей передаем, а оставшиеся после сортировки значения объединяет в 'other':

```python
def top_statistic_v2(df, groupby_column, agg_column):
    """
    df - датафрейм
    groupby_column - колонка, по которой происходит группировка
    agg_column - колонка по которой происходит агрегирование
    """
    df_top = (
        df
        .groupby(groupby_column)
        .agg({agg_column: 'sum'})
        .sort_values(by=agg_column, ascending=False)
    )
    top_index = df_top.head().index.tolist()
    df_top.index = df_top.index.where(df_top.index.isin(top_index), 'other')
    df_top = (
        df_top
        .groupby(groupby_column)
        .agg({agg_column: 'sum'})
        .sort_values(by=agg_column, ascending=False)
    )
    return df_top
```

```python
top_genre_na = top_statistic_v2(actual_games, 'genre', 'na_sales')
top_genre_eu = top_statistic_v2(actual_games, 'genre', 'eu_sales')
top_genre_jp = top_statistic_v2(actual_games, 'genre', 'jp_sales')
```

## Проверка гипотез

Будем использовать t-критерий Стьюдента, так как сравниваемые выборки независимы между собой

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Сформулируем гипотезы (нулевая гипотеза Н0 формулируется со знаком равенства. Исходя из нулевой гипотезы, формируется альтернативная гипотеза Н1, которая противоречит нулевой):
- H0: Средние пользовательские рейтинги платформ Xbox One и PC равны
- H1: Средние пользовательские рейтинги платформ Xbox One и PC не равны

```python
XOne  = actual_games.query('platform == "XOne" and user_score != -99')['user_score']
PC  = actual_games.query('platform == "PC" and user_score != -99')['user_score']

print('Дисперсия выборки рейтинга платформы "XOne":', np.var(XOne))
print('Дисперсия выборки рейтинга платформы "PC":', np.var(PC))
```

Дисперсии выборок не равны, поэтому будем использовать параметр `equal_var` = `False`

```python
#примем критический уровень статистической значимости alpha равный 0.05
alpha = .05

results = st.ttest_ind(XOne, PC, equal_var=False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Вывод**:

Значение p-value превышает уровень значимости, примем нулевую гипотезу: "Средние пользовательские рейтинги платформ Xbox One и PC равны"

```python
(
    actual_games
    .query('platform in ["XOne", "PC"] and user_score != -99')
    .pivot_table(index='platform', values='user_score', aggfunc='mean')
)
```

### Средние пользовательские рейтинги жанров Action и Sports разные

Сформулируем гипотезы:
- H0: средние пользовательские рейтинги жанров Action и Sports равны
- H1: средние пользовательские рейтинги жанров Action и Sports не равны

```python
Action  = actual_games.query('genre == "Action" and user_score != -99')['user_score']
Sports  = actual_games.query('genre == "Sports" and user_score != -99')['user_score']

print('Дисперсия выборки рейтинга жанра "Action":', np.var(Action))
print('Дисперсия выборки рейтинга жанра "Sports":', np.var(Sports))
```

Дисперсии выборок не равны, поэтому будем использовать параметр `equal_var` = `False`

```python
#примем критический уровень статистической значимости alpha равный 0.05
alpha = .05

results = st.ttest_ind(Action, Sports, equal_var=False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

5.98 на 10 в -20 степени, какой бы уровень значимости мы бы не выдвигали, p-value в любом случае туда не поместится, так что отвергнем нулевую гипотезу

```python
(
    actual_games
    .query('genre in ["Action", "Sports"] and user_score != -99')
    .pivot_table(index='genre', values='user_score', aggfunc='mean')
)
```

Средние значения рейтингов двух жанров различаются
